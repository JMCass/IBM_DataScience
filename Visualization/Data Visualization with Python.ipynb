# Matplotlib

```python
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas #importar FigureCanvas
from matplotlib.figure import Figure #importar artista Figure
fig = Figure()
canvas = FigureCanvas(fig)

import numpy as np
x = np.random.randn(10000)

ax = fig.add_subplot(111) #crear un Eje artista

ax.hist(x, 100) #generar un histograma de 10000 n√∫meros

ax.set_title('Normal distribution with $\mu=0, \sigma=1$')

fig.savefig('matplotlib_histogram.png')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(10000)
plt.hist(x,100)
plt.title(r'Normal distribution with $\mu=0, \sigma=1$')
plt.savefig('matplotlib_histogram2.png')
plt.show();
```

```python
plt.plot(5,5, 'o')
plt.title('Example using matplotlib')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show();
```

```python
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
```

```python
df_can.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df_can.columns
```

```python
df_can['Total'] = df_can.sum(axis=1)
```

```python
df_can.set_index('Country', inplace=True)
```

```python
years = list(map(int,range(1980, 2014)))
```

```python
haiti = df_can.loc['Haiti', years]
mexico = df_can.loc['Mexico', years]
```

```python
haiti.index = haiti.index.map(int)
mexico = df_can.loc['Mexico', years]

haiti.plot(kind='line')
mexico.plot(kind='line')
plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.show() # need this line to show the updates made to the figure
```
